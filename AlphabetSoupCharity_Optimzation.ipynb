## First Attempt
Below you will see my first attempt to raiss the model score: I will change the binning threshold numbers to create more features in the models and perhaps increase the score.

```python
application_df1["APPLICATION_TYPE"].value_counts()
```

```python
application_df1["CLASSIFICATION"].value_counts()
```

```python
'''
below if the original code that puts rare rows into categoriesin the APPLICATION_TYPE column. <= 1200 was the original number I will change it to 
500 resulting in more features after one-hot encoding
'''

type_df = application_df1.apply(lambda x: x.mask(x.map(x.value_counts())<=500, 'other') if x.name=='APPLICATION_TYPE' else x)
type_df
```

```python
'''
In the original code that puts rare rows into categories 
in the Classification column, was <= 4000 
was the original number. I will change it to 
7000 resulting in more features after one-hot encoding
'''

application_df2 = type_df.apply(lambda x: x.mask(x.map(x.value_counts())<=4000, 'other') if x.name=='CLASSIFICATION' else x)


```

In the previous notebook the one-hot encoding reulted in 36 features. below there is an increase in columns to 42.

```python
numeric_df = pd.get_dummies(application_df2)
numeric_df
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = numeric_df["IS_SUCCESSFUL"]
X = numeric_df.drop("IS_SUCCESSFUL", axis=1)



# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn_model = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE

nn_model.add(tf.keras.layers.Dense(units=15, activation="relu", input_dim=41))

# Second hidden layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=25, activation="relu"))

# Output layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE

nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn_model.fit(X_train_scaled,y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Second attempt
